##Find if an element exists in a list.

```python
arr = [10, 20, 30, 40]   # sample input
key = 30

found = False

for num in arr:
    if num == key:
        found = True
        break

if found:
    print("Found")
else:
    print("Not Found")
```

## Binary Search (Important)Works only on sorted list

```python
arr = [1, 2, 3, 4, 5]   # sorted list
key = 4

low = 0
high = len(arr) - 1
found = False

while low <= high:
    mid = (low + high) // 2
    
    if arr[mid] == key:
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1

print("Found" if found else "Not Found")
```

##Bubble Sort

```python
arr = [5, 2, 9, 1]   # sample input

n = len(arr)

for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print(arr)
```

##Find Largest and Smallest in List

```python
arr = [7, 2, 9, 4, 1]   # sample input

small = large = arr[0]

for num in arr:
    if num < small:
        small = num
    if num > large:
        large = num

print("Smallest:", small)
print("Largest:", large)
```

##Selection Sort

```python
arr = [64, 25, 12, 22, 11]   # sample input

n = len(arr)

for i in range(n):
    min_idx = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_idx]:
            min_idx = j
    
    arr[i], arr[min_idx] = arr[min_idx], arr[i]

print(arr)
```
