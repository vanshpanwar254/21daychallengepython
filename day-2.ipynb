##Write a program to reverse a given number.

```python
n = 1234   # sample input
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print(rev)
```

##Write a program to check whether a number is a palindrome or not.

```python
n = 121   # sample input
temp = n
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

##Write a program to count how many digits are in a number.

```python
n = 12345   # sample input
count = 0

while n > 0:
    n = n // 10
    count += 1

print(count)
```

##Print the following pattern:

```python
n = 5   # sample input

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```
