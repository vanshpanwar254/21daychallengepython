##Find Second Largest Element in List

```python
arr = [10, 20, 4, 45, 99]  

first = second = float('-inf')

for num in arr:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num

print(second)
```

##Reverse a List (Without using reverse())

```python
arr = [1, 2, 3, 4]   

rev = []

for i in range(len(arr)-1, -1, -1):
    rev.append(arr[i])

print(rev)
```

##Sum of All Elements in List

```python
arr = [5, 10, 15]  

total = 0

for num in arr:
    total += num

print(total)
```

##Write a program to find common elements between two lists (without using built-in set intersection).

```python
arr1 = [1, 2, 3, 4]   
arr2 = [3, 4, 5, 6]   

common = []

for num in arr1:
    if num in arr2 and num not in common:
        common.append(num)

print(common)
```
