##Find two numbers whose sum = target (use efficient approach).

```python
arr = [2, 7, 11, 15]   # sample input
target = 9

seen = {}

for num in arr:
    diff = target - num
    if diff in seen:
        print((diff, num))
        break
    seen[num] = True
```

##Move All Zeros to End

```python
arr = [1, 2, 3, 4]   # sample input

rev = []

for i in range(len(arr)-1, -1, -1):
    rev.append(arr[i])

print(rev)
```

##Write a program to find common elements between two lists (without using built-in set intersection).

```python
arr1 = [1, 2, 3, 4]   # sample input
arr2 = [3, 4, 5, 6]   # sample input

common = []

for num in arr1:
    if num in arr2 and num not in common:
        common.append(num)

print(common)
```

##Rotate List to Right by K Steps

```python
arr = [1, 2, 3, 4, 5]   # sample input
k = 2

k = k % len(arr)   # handle large k

result = arr[-k:] + arr[:-k]

print(result)
```

##Check if String Contains Only Digits

```python
s = "12345"   # sample input

is_digit = True

for ch in s:
    if ch < '0' or ch > '9':
        is_digit = False
        break

print(is_digit)
```
