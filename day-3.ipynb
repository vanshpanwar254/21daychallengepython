## Write a recursive program to find the sum of first n natural numbers.

```python
def natural_sum(n):
    if n == 1:
        return 1
    else:
        return n + natural_sum(n - 1)

print("Sum =", natural_sum(6))
```

##Count number of vowels in a string.

```python
s = "education"   
count = 0

for ch in s:
    if ch in "aeiouAEIOU":
        count += 1

print(count)
```

##Remove all spaces from a string.

```python
s = "hello world"   
result = ""

for ch in s:
    if ch != " ":
        result += ch

print(result)
```

##Find Maximum Element in a List

```python
arr = [3, 7, 2, 9, 5]   

max_val = arr[0]

for num in arr:
    if num > max_val:
        max_val = num

print(max_val)
```

##Remove Duplicates from List

```python
arr = [1, 2, 2, 3, 4, 4, 5] 
result = []

for num in arr:
    if num not in result:
        result.append(num)

print(result)
```
