##Find Frequency of Characters in String

```python
s = "aabbc"  
freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

##Find Missing Number in List (1 to N)

```python
arr = [1, 2, 4, 5]   

n = len(arr) + 1
total = n * (n + 1) // 2
actual = sum(arr)

print(total - actual)
```

##Find All Pairs with Given Sum

```python
arr = [1, 2, 3, 4, 5]   
target = 6

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] + arr[j] == target:
            print((arr[i], arr[j]))
```

##Capitalize First Letter of Each Word

```python
s = "hello world"  
result = ""

words = s.split()

for w in words:
    result += w[0].upper() + w[1:] + " "

print(result.strip())
```

##Find Duplicate Elements in List

```python
arr = [1, 2, 3, 2, 4, 1, 5]   

duplicates = []

for num in arr:
    if arr.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

print(duplicates)
```
