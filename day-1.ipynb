##Write a program to take a number from the user and check whether it is even or odd.

```python
n = 7   # sample input

if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

##Write a program to take three integers as input from the user and print the largest among them.

```python
a, b, c = 3, 9, 5   

if a >= b and a >= c:
    print(a)
elif b >= a and b >= c:
    print(b)
else:
    print(c)
```

##Write a program to take an integer N as input and print numbers from 1 to N.

```python
n = 5  

for i in range(1, n + 1):
    print(i, end=" ")
```

##Write a program to take an integer N as input and print the sum of first N natural numbers.

```python
n = 5   

total = n * (n + 1) // 2
print(total)
```
