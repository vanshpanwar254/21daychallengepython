##Find GCD of Two Numbers

```python
a = 12   # sample input
b = 18

while b != 0:
    a, b = b, a % b

print(a)
```

##Flatten a Nested List

```python
arr = [1, [2, 3], [4, 5]]   # sample input

result = []

for item in arr:
    if isinstance(item, list):
        for sub in item:
            result.append(sub)
    else:
        result.append(item)

print(result)
```

##Find Second Smallest Element

```python
arr = [5, 2, 8, 1, 3]   # sample input

unique = list(set(arr))
unique.sort()

print(unique[1])
```

##Check if List is Sorted

```python
arr = [1, 2, 3, 4, 5]   # sample input

is_sorted = True

for i in range(len(arr) - 1):
    if arr[i] > arr[i + 1]:
        is_sorted = False
        break

print(is_sorted)
```
